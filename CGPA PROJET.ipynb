```python
import csv
# Registration Form
registration_form = {
    "names": [],
    "email": "",
    "password": "",
    "gender":"",
    "department":"",
    "level":"",
    "programe":"",
    "confirm_password": "",
    "session": "",
    "matric": "",
    "first_semester_course_titles": [],
    "first_semester_course_codes": [],
    "first_semester_course_units": [],
    "first_semester_course_scores": [],
    "second_semester_course_titles": [],
    "second_semester_course_codes": [],
    "second_semester_course_units": [],
    "second_semester_course_scores": []
}

# Input personal information
print("Registration Form")
print("------------------")

# Validate and append name
while True:
    name = input("Enter your name: ")
    if name.replace(" ", "").isalpha():
        registration_form["names"].append(name)
        break
    else:
        print("Invalid name. Please try again.")

# Validate and set email
while True:
    email = input("Enter your email: ")
    if "@" in email and email.endswith(".com"):
        registration_form["email"] = email
        break
    else:
        print("Invalid email. Please try again.")

# Validate and set password
while True:
    password = input("Enter your password: ")
    if password.isalnum() and len(password) >= 6:
        registration_form["password"] = password
        break
    else:
         print("Passwords do not match. Please try again.")

        
        

    #registration_form["password"] = input("Enter your password: ")
    #registration_form["confirm_password"] = input("Confirm your password: ")
    #if registration_form["password"] == registration_form["confirm_password"]:
        

# Validate and set session
while True:
    session = input("Enter your session: ")
    if session == "2023/2024":
        registration_form["session"] = session
        break
    else:
        print("Invalid session. Please try again.")
# set level
while True:
    level = input("Enter your level(100/200/300/400/500): ")
    if level.isdigit() and level in ["100","200","300","400","500"]:
        registration_form["level"] = level
        break
    else:
        print("Invald level.Please try again.")

# Set matric
while True:
    matric = input("Enter your matric number: ")
    if matric.isdigit():
        registration_form["matric"] = matric
        break
    else:
        print("Invalid matric number. Please try again.")
while True:
    gender = input("Enter your gender(male/female): ")
    if gender.lower() in ["male","female"]:
        registration_form["gender"] = gender
        break
    else:
        print("invalid gender.Please try again.")
while True:
    department = input("Enter your department: ")
    if department.replace(" ","").isalpha():
        registration_form["department"] = department
        break
    else:
        print("invalid department.Please try again.")
while True:
    programe = input("Enter your programe(b.tech/bsc")
    if programe.lower() in ["b.tech","bsc"]:
        registration_form["programe"] = programe
        break
    else:
        print("Invalid programe.Please try again.")

        
# Function to input courses with a limit of 6
def input_courses(semester):
    print(f"\nEnter courses for the {semester} (up to 6 courses):")
    titles = []
    codes = []
    units = []
    
    while len(titles) < 6:
        course_title = input("Enter course title: ")
        titles.append(course_title)

        course_code = input("Enter course code: ")
        codes.append(course_code)

        units.append(int(input(f"Enter number of units for {course_code}: ")))

        if len(titles) < 6:
            continue_choice = input("Do you want to add another course? (yes/no): ").strip().lower()
            if continue_choice != 'yes':
                break
        else:
            print("You have reached the maximum number of courses for this semester.")

    return titles, codes, units

# Input first semester courses
registration_form["first_semester_course_titles"], registration_form["first_semester_course_codes"], registration_form["first_semester_course_units"] = input_courses("first semester")

# Input second semester courses
registration_form["second_semester_course_titles"], registration_form["second_semester_course_codes"], registration_form["second_semester_course_units"] = input_courses("second semester")

# Function to calculate CGPA and store scores
def calculate_cgpa(course_titles, course_codes, course_units):
    if not course_titles:
        return 0, 0, 0, []

    scores = []
    print("\nEnter scores for the courses:")
    for code in course_codes:
        score = float(input(f"Enter your score for {code}: "))
        scores.append(score)

    total_units = 0
    total_points = 0

    print("\nGrades and CGPA:")
    for i in range(len(course_titles)):
        title = course_titles[i]
        code = course_codes[i]
        units = course_units[i]
        score = scores[i]

        if score >= 70:
            grade = "A"
            point = 5
        elif score >= 60:
            grade = "B"
            point = 4
        elif score >= 50:
            grade = "C"
            point = 3
        elif score >= 45:
            grade = "D"
            point = 2
        elif score >= 40:
            grade = "E"
            point = 1
        else:
            grade = "F"
            point = 0

        print(f"{title} ({code}): {score} ({grade}) - {point} point(s)")
        total_units += units
        total_points += point * units

    cgpa = total_points / total_units if total_units != 0 else 0
    return total_units, total_points, cgpa, scores

# Calculate first semester results
first_total_units, first_total_points, first_cgpa, registration_form["first_semester_course_scores"] = calculate_cgpa(
    registration_form["first_semester_course_titles"],
    registration_form["first_semester_course_codes"],
    registration_form["first_semester_course_units"]
)

# Calculate second semester results
second_total_units, second_total_points, second_cgpa, registration_form["second_semester_course_scores"] = calculate_cgpa(
    registration_form["second_semester_course_titles"],
    registration_form["second_semester_course_codes"],
    registration_form["second_semester_course_units"]
)

# Calculate and display overall CGPA
combined_total_units = first_total_units + second_total_units
combined_total_points = first_total_points + second_total_points

if combined_total_units != 0:
    combined_cgpa = combined_total_points / combined_total_units
    overall_class = ""
    if combined_cgpa >= 4.5:
        overall_class = "First class"
    elif combined_cgpa >= 3.50:
        overall_class = "Second class upper"
    elif combined_cgpa >= 2.40:
        overall_class = "Second class lower"
    elif combined_cgpa >= 1.50:
        overall_class = "Third class"
    elif combined_cgpa >= 1.00:
        overall_class = "Pass"
    else:
        overall_class = "Fail"
else:
    combined_cgpa = 0
    overall_class = "No courses registered."

# Display final registration information
def show_table():
    first_semester_courses = len(registration_form["first_semester_course_titles"])
    second_semester_courses = len(registration_form["second_semester_course_titles"])
    if first_semester_courses == 1 and second_semester_courses == 1:
        print(f"""
        **********************************************************************************************************************
        *                                         DEPARTMENT OF STATISTICS                                                  *
        *                                  LADOKE AKINTOLA UNIVERSITY OF TECHNOLOGY,                                         *
        *                                    P.M.B 4000, OGBOMOSHO, OYO STATE.                                               *
        **********************************************************************************************************************

        ********************************************************************************************************************
        *                                       Course Registration Form                                                  *         
        ********************************************************************************************************************

*****************************************          *****************************************     **************************************************
* MATRIC:{registration_form['matric']} |           * SESSION:{registration_form['session']} |    * NAME(S): {', '.join(registration_form['names'])}|
****************************************           *****************************************     ***************************************************

***************************************           *******************************************     ***********************************************
* LEVEL : {registration_form['level']} |          * Programe:{registration_form['programe']} |    * Department:{registration_form['department']}|
****************************************          ********************************************    ***********************************************

****************************************
* Gender: {registration_form['gender']}|
****************************************


                                     *****************************************************************
                                     *                First Semester Result                          *
                                     *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     | {registration_form['first_semester_course_titles'][0]}| {registration_form['first_semester_course_codes'][0]} | {registration_form['first_semester_course_units'][0]} | {registration_form['first_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

***************************************
First Semester GPA: {first_cgpa:.2f}")|
***************************************






                                     *****************************************************************
                                     *                Second Semester Result                          *
                                     *****************************************************************

___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     |{registration_form['second_semester_course_titles'][0]}|{registration_form['second_semester_course_codes'][0]} |{registration_form['second_semester_course_units'][0]} |{registration_form['second_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

***************************************
Second Semester GPA: {second_cgpa:.2f}|
***************************************

***********************************
Overall CGPA: {combined_cgpa:.2f} |
***********************************

***********************
Class: {overall_class}|
***********************

""")

    
    elif first_semester_courses == 2 and second_semester_courses == 2:
        print(f"""
             **********************************************************************************************************************
             *                                         DEPARTMENT OF STATISTICS                                                  *
             *                                  LADOKE AKINTOLA UNIVERSITY OF TECHNOLOGY,                                         *
             *                                    P.M.B 4000, OGBOMOSHO, OYO STATE.                                               *
             **********************************************************************************************************************

             ********************************************************************************************************************
             *                                       Course Registration Form                                                  *         
             ********************************************************************************************************************

*****************************************          *****************************************     **************************************************
* MATRIC:{registration_form['matric']} |           * SESSION:{registration_form['session']} |    * NAME(S): {', '.join(registration_form['names'])}|
****************************************           *****************************************     ***************************************************

***************************************           *******************************************     ***********************************************
* LEVEL : {registration_form['level']} |          * Programe:{registration_form['programe']} |    * Department:{registration_form['department']}|
****************************************          ********************************************    ***********************************************

****************************************
* Gender: {registration_form['gender']}|
****************************************

                                    *****************************************************************
                                    *                First Semester Result                          *
                                     *****************************************************************
 ___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     | {registration_form['first_semester_course_titles'][0]}| {registration_form['first_semester_course_codes'][0]} | {registration_form['first_semester_course_units'][0]} | {registration_form['first_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     | {registration_form['first_semester_course_titles'][1]}| {registration_form['first_semester_course_codes'][1]} | {registration_form['first_semester_course_units'][1]} | {registration_form['first_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

***************************************
First Semester GPA: {first_cgpa:.2f}")|
***************************************
                                    
        
                                *****************************************************************
                                *                Second Semester Result                         *
                                *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     |{registration_form['second_semester_course_titles'][0]}|{registration_form['second_semester_course_codes'][0]} | {registration_form['first_semester_course_units'][0]} |{registration_form['second_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     |{registration_form['second_semester_course_titles'][1]}|{registration_form['second_semester_course_codes'][1]} | {registration_form['first_semester_course_units'][1]} |{registration_form['second_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

****************************************
Second Semester GPA: {second_cgpa:.2f}|
***************************************

**********************************
Overall CGPA: {combined_cgpa:.2f}|
**********************************

***********************
Class: {overall_class}|
***********************



""")
    elif first_semester_courses == 3 and second_semester_courses == 3:
        print(f"""
**********************************************************************************************************************
*                                         DEPARTMENT OF STATISTICS                                                  *
*                                  LADOKE AKINTOLA UNIVERSITY OF TECHNOLOGY,                                         *
*                                    P.M.B 4000, OGBOMOSHO, OYO STATE.                                               *
**********************************************************************************************************************

********************************************************************************************************************
*                                       Course Registration Form                                                  *         
********************************************************************************************************************

*****************************************          *****************************************     **************************************************
* MATRIC:{registration_form['matric']} |           * SESSION:{registration_form['session']} |    * NAME(S): {', '.join(registration_form['names'])}|
****************************************           *****************************************     ***************************************************

***************************************           *******************************************     ***********************************************
* LEVEL : {registration_form['level']} |          * Programe:{registration_form['programe']} |    * Department:{registration_form['department']}|
****************************************          ********************************************    ***********************************************

****************************************
* Gender: {registration_form['gender']}|
****************************************
                                    *****************************************************************
                                    *                First Semester Result                          *
                                     *****************************************************************

___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     | {registration_form['first_semester_course_titles'][0]}| {registration_form['first_semester_course_codes'][0]} | {registration_form['first_semester_course_units'][0]} | {registration_form['first_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     | {registration_form['first_semester_course_titles'][1]}| {registration_form['first_semester_course_codes'][1]} | {registration_form['first_semester_course_units'][1]} | {registration_form['first_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     | {registration_form['first_semester_course_titles'][2]}|{registration_form['first_semester_course_codes'][2]}  | {registration_form['first_semester_course_units'][2]} |{registration_form['first_semester_course_scores'][2]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

****************************************
First Semester GPA: {first_cgpa:.2f}") |
****************************************                                    
        
                                *****************************************************************
                                *                Second Semester Result                         *
                                *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     |{registration_form['second_semester_course_titles'][0]}|{registration_form['second_semester_course_codes'][0]} |{registration_form['second_semester_course_units'][0]} |{registration_form['second_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     |{registration_form['second_semester_course_titles'][1]}|{registration_form['second_semester_course_codes'][1]} |{registration_form['second_semester_course_units'][1]} |{registration_form['second_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     |{registration_form['second_semester_course_titles'][2]}|{registration_form['second_semester_course_codes'][2]} |{registration_form['second_semester_course_units'][2]} |{registration_form['second_semester_course_scores'][2]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

****************************************
Second Semester GPA: {second_cgpa:.2f}
***************************************

**********************************
Overall CGPA: {combined_cgpa:.2f}|
********************************

***********************
Class: {overall_class}|
***********************

""")
    elif first_semester_courses == 4 and second_semester_courses == 4:
         print(f"""
                        **********************************************************************************************************************
                        *                                         DEPARTMENT OF STATISTICS                                                  *
                        *                                  LADOKE AKINTOLA UNIVERSITY OF TECHNOLOGY,                                         *
                        *                                    P.M.B 4000, OGBOMOSHO, OYO STATE.                                               *
                        **********************************************************************************************************************

                        ********************************************************************************************************************
                        *                                       Course Registration Form                                                  *         
                        ********************************************************************************************************************

*****************************************          *****************************************     **************************************************
* MATRIC:{registration_form['matric']} |           * SESSION:{registration_form['session']} |    * NAME(S): {', '.join(registration_form['names'])}|
****************************************           *****************************************     ***************************************************

***************************************           *******************************************     ***********************************************
* LEVEL : {registration_form['level']} |          * Programe:{registration_form['programe']} |    * Department:{registration_form['department']}|
****************************************          ********************************************    ***********************************************

****************************************
* Gender: {registration_form['gender']}|
****************************************
                                    *****************************************************************
                                    *                First Semester Result                          *
                                     *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     | {registration_form['first_semester_course_titles'][0]}| {registration_form['first_semester_course_codes'][0]} | {registration_form['first_semester_course_units'][0]} | {registration_form['first_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     | {registration_form['first_semester_course_titles'][1]}| {registration_form['first_semester_course_codes'][1]} | {registration_form['first_semester_course_units'][1]} | {registration_form['first_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     | {registration_form['first_semester_course_titles'][2]}|{registration_form['first_semester_course_codes'][2]}  | {registration_form['first_semester_course_units'][2]} |{registration_form['first_semester_course_scores'][2]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   4     | {registration_form['first_semester_course_titles'][3]}|{registration_form['first_semester_course_codes'][3]}  |{registration_form['first_semester_course_units'][3]}  |{registration_form['first_semester_course_scores'][3]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

***************************************
First Semester GPA: {first_cgpa:.2f}")|
***************************************
                                    
                                    
                                    *****************************************************************
                                    *                Second Semester Result                          *
                                    *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     |{registration_form['second_semester_course_titles'][0]}|{registration_form['second_semester_course_codes'][0]} |{registration_form['second_semester_course_units'][0]} |{registration_form['second_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     |{registration_form['second_semester_course_titles'][1]}|{registration_form['second_semester_course_codes'][1]} |{registration_form['second_semester_course_units'][1]} |{registration_form['second_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     |{registration_form['second_semester_course_titles'][2]}|{registration_form['second_semester_course_codes'][2]} |{registration_form['second_semester_course_units'][2]} |{registration_form['second_semester_course_scores'][2]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   4     |{registration_form['second_semester_course_titles'][3]}|{registration_form['second_semester_course_codes'][3]} |{registration_form['second_semester_course_units'][3]} |{registration_form['second_semester_course_scores'][3]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

***************************************
Second Semester GPA: {second_cgpa:.2f}|
***************************************

***********************************
Overall CGPA: {combined_cgpa:.2f} |
***********************************

************************
Class: {overall_class} |
************************
""")
    elif first_semester_courses == 5 and second_semester_courses == 5:
         print(f"""
                    **********************************************************************************************************************
                    *                                         DEPARTMENT OF STATISTICS                                                  *
                    *                                  LADOKE AKINTOLA UNIVERSITY OF TECHNOLOGY,                                         *
                    *                                    P.M.B 4000, OGBOMOSHO, OYO STATE.                                               *
                    **********************************************************************************************************************

                    ********************************************************************************************************************
                    *                                       Course Registration Form                                                  *         
                    ********************************************************************************************************************

*****************************************          *****************************************     **************************************************
* MATRIC:{registration_form['matric']} |           * SESSION:{registration_form['session']} |    * NAME(S): {', '.join(registration_form['names'])}|
****************************************           *****************************************     ***************************************************

***************************************           *******************************************     ***********************************************
* LEVEL : {registration_form['level']} |          * Programe:{registration_form['programe']} |    * Department:{registration_form['department']}|
****************************************          ********************************************    ***********************************************

****************************************
* Gender: {registration_form['gender']}|
****************************************
                                    *****************************************************************
                                    *                First Semester Result                          *
                                    *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     | {registration_form['first_semester_course_titles'][0]}| {registration_form['first_semester_course_codes'][0]} | {registration_form['first_semester_course_units'][0]} | {registration_form['first_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     | {registration_form['first_semester_course_titles'][1]}| {registration_form['first_semester_course_codes'][1]} | {registration_form['first_semester_course_units'][1]} | {registration_form['first_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     | {registration_form['first_semester_course_titles'][2]}|{registration_form['first_semester_course_codes'][2]}  | {registration_form['first_semester_course_units'][2]} |{registration_form['first_semester_course_scores'][2]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   4     | {registration_form['first_semester_course_titles'][3]}|{registration_form['first_semester_course_codes'][3]}  |{registration_form['first_semester_course_units'][3]}  |{registration_form['first_semester_course_scores'][3]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   5     | {registration_form['first_semester_course_titles'][4]}|{registration_form['first_semester_course_codes'][4]}  |{registration_form['first_semester_course_units'][4]}  |{registration_form['first_semester_course_scores'][4]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

***************************************
First Semester GPA: {first_cgpa:.2f}")|
***************************************


                                    *****************************************************************
                                    *                Second Semester Result                          *
                                    *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     |{registration_form['second_semester_course_titles'][0]}|{registration_form['second_semester_course_codes'][0]} |{registration_form['second_semester_course_units'][0]} |{registration_form['second_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     |{registration_form['second_semester_course_titles'][1]}|{registration_form['second_semester_course_codes'][1]} |{registration_form['second_semester_course_units'][1]} |{registration_form['second_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     |{registration_form['second_semester_course_titles'][2]}|{registration_form['second_semester_course_codes'][2]} |{registration_form['second_semester_course_units'][2]} |{registration_form['second_semester_course_scores'][2]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   4     |{registration_form['second_semester_course_titles'][3]}|{registration_form['second_semester_course_codes'][3]} |{registration_form['second_semester_course_units'][3]} |{registration_form['second_semester_course_scores'][3]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   5     |{registration_form['second_semester_course_titles'][4]}|{registration_form['second_semester_course_codes'][4]} |{registration_form['second_semester_course_units'][4]} |{registration_form['second_semester_course_scores'][4]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|    

***************************************
Second Semester GPA: {second_cgpa:.2f}|
***************************************

**********************************
Overall CGPA: {combined_cgpa:.2f}|
**********************************

************************
Class: {overall_class} |
************************

""")
    elif first_semester_courses == 6 and second_semester_courses == 6:
        print(f"""
                    **********************************************************************************************************************
                    *                                         DEPARTMENT OF STATISTICS                                                  *
                    *                                  LADOKE AKINTOLA UNIVERSITY OF TECHNOLOGY,                                         *
                    *                                    P.M.B 4000, OGBOMOSHO, OYO STATE.                                               *
                    **********************************************************************************************************************

                    ********************************************************************************************************************
                    *                                       Course Registration Form                                                  *         
                    ********************************************************************************************************************

*****************************************          *****************************************     **************************************************
* MATRIC:{registration_form['matric']} |           * SESSION:{registration_form['session']} |    * NAME(S): {', '.join(registration_form['names'])}|
****************************************           *****************************************     ***************************************************

***************************************           *******************************************     ***********************************************
* LEVEL : {registration_form['level']} |          * Programe:{registration_form['programe']} |    * Department:{registration_form['department']}|
****************************************          ********************************************    ***********************************************

****************************************
* Gender: {registration_form['gender']}|
****************************************
                                    *****************************************************************
                                    *                First Semester Result                          *
                                    *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     | {registration_form['first_semester_course_titles'][0]}| {registration_form['first_semester_course_codes'][0]} | {registration_form['first_semester_course_units'][0]} | {registration_form['first_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     | {registration_form['first_semester_course_titles'][1]}| {registration_form['first_semester_course_codes'][1]} | {registration_form['first_semester_course_units'][1]} | {registration_form['first_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     | {registration_form['first_semester_course_titles'][2]}|{registration_form['first_semester_course_codes'][2]}  | {registration_form['first_semester_course_units'][2]} |{registration_form['first_semester_course_scores'][2]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   4     | {registration_form['first_semester_course_titles'][3]}|{registration_form['first_semester_course_codes'][3]}  |{registration_form['first_semester_course_units'][3]}  |{registration_form['first_semester_course_scores'][3]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   5     | {registration_form['first_semester_course_titles'][4]}|{registration_form['first_semester_course_codes'][4]}  |{registration_form['first_semester_course_units'][4]}  |{registration_form['first_semester_course_scores'][4]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   6     |{registration_form['first_semester_course_titles'][5]} |{registration_form['first_semester_course_codes'][5]}  |{registration_form['first_semester_course_units'][5]}  |{registration_form['first_semester_course_scores'][4]} |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

*****************************************
First Semester GPA: {first_cgpa:.2f}")  |
*****************************************


                                    *****************************************************************
                                    *                Second Semester Result                          *
                                    *****************************************************************
___________________________________________________________________________________________________________________________________________________________________________________________________________________________________________
|  S/N    |                  Course Titles                        |      Course Code                                      |      Course Unit                                      |  Score                                                |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   1     |{registration_form['second_semester_course_titles'][0]}|{registration_form['second_semester_course_codes'][0]} |{registration_form['second_semester_course_units'][0]} |{registration_form['second_semester_course_scores'][0]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   2     |{registration_form['second_semester_course_titles'][1]}|{registration_form['second_semester_course_codes'][1]} |{registration_form['second_semester_course_units'][1]} |{registration_form['second_semester_course_scores'][1]}|                                                  |
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   3     |{registration_form['second_semester_course_titles'][2]}|{registration_form['second_semester_course_codes'][2]} |{registration_form['second_semester_course_units'][2]} |{registration_form['second_semester_course_scores'][2]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   4     |{registration_form['second_semester_course_titles'][3]}|{registration_form['second_semester_course_codes'][3]} |{registration_form['second_semester_course_units'][3]} |{registration_form['second_semester_course_scores'][3]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|
|   5     |{registration_form['second_semester_course_titles'][4]}|{registration_form['second_semester_course_codes'][4]} |{registration_form['second_semester_course_units'][4]} |{registration_form['second_semester_course_scores'][4]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________| 
|   6     |{registration_form['second_semester_course_titles'][5]}|{registration_form['second_semester_course_codes'][5]} |{registration_form['second_semester_course_units'][5]} |{registration_form['second_semester_course_scores'][5]}|
|_________|_______________________________________________________|_______________________________________________________|_______________________________________________________|_______________________________________________________|

***************************************
Second Semester GPA: {second_cgpa:.2f} |
****************************************

***********************************
Overall CGPA: {combined_cgpa:.2f} |
**********************************

************************
Class: {overall_class} |
************************


""")

show_table()

# Write registration form data to CSV file
with open('registration_formss.csv', mode = "a") as f:
    writer = csv.writer(f)
    writer.writerow(['Matric', 'Session', 'Name', 'Email', 'Password', 'First Semester Courses', 'First Semester Scores', 'Second Semester Courses', 'Second Semester Scores', 'Overall CGPA', 'Class'])
    writer.writerow([registration_form['matric'], registration_form['session'], ', '.join(registration_form['names']), registration_form['email'], registration_form['password'], 
                     ', '.join(registration_form['first_semester_course_titles']), ', '.join(map(str, registration_form['first_semester_course_scores'])), 
                     ', '.join(registration_form['second_semester_course_titles']), ', '.join(map(str, registration_form['second_semester_course_scores'])), 
                     combined_cgpa, overall_class])
```
